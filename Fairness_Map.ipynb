##### Function to check if two precinct overlap (needs precinct_boundaries to be instanciated)

```python
def is_contiguous_precinct(p1,p2,precinct_boundaries):
    #print(shape(precinct_boundaries[p1]).type)
    try:
        if shape(precinct_boundaries[p1]).type == 'Polygon':
            a = Polygon(shape(precinct_boundaries[p1]))
        else:
            return False
        if shape(precinct_boundaries[p2]).type == 'Polygon':
            b = Polygon(shape(precinct_boundaries[p2]))
        else:
            return False
        return(a.touches(b))
    except KeyError: return False
```

```python
print(is_contiguous_precinct('34003060003','34007043046',precinct_boundaries))
print(is_contiguous_precinct('34007043047','34007043046',precinct_boundaries))
print(is_contiguous_precinct('34003060003','34003060002',precinct_boundaries))
print(is_contiguous_precinct('34007043047','34003060003',precinct_boundaries))
print(is_contiguous_precinct('34041080003','34005020001',precinct_boundaries))
print(is_contiguous_precinct('34041080003','34005075101',precinct_boundaries))
```

##### Function to find contiguous districts (needs precinct_boundaries to be instanciated)

```python
## This code will tell you how many precinct are contiguous to a given precinct
def contiguous_precincts(p1,precinct_boundaries):
    count=0
    neighbors=[]
    for p in range(0,len(pre_data)):
        #print(pre_data.iloc[p].GEOID20)
        if(is_contiguous_precinct(p1,pre_data.iloc[p].GEOID20,precinct_boundaries)):
            count+=1
            neighbors.append(pre_data.iloc[p].GEOID20)
    return([count,neighbors])
```

```python
def sortSecond(val):
    return val[1]
```

```python
def district_partisan_lean(dataframe):
    assignment = dataframe
    support=0
    sum_support=0
    district_support = []
    i = 0
    for districts in (assignment.groupby(by="District")):
        #print(districts)
        
        for precinct in districts[1].iterrows():
            #print(precinct[1][0])
            
            if(pre_data.loc[pre_data.GEOID20==str(precinct[1][0]),'Dem_2016-2021_Comp'].any()):
                support += pre_data.loc[pre_data.GEOID20==str(precinct[1][0]),'Dem_2016-2021_Comp'].values[0]
        
        
        print("The democratic support of District "+str(districts[0])+ " is "+str(support))
        district_support = district_support + [(str(districts[0]), str(support))]
        sum_support+=support
        support=0
        
        
    print("Total democratic support is "+str(sum_support))
    district_support.sort(key=sortSecond, reverse=True)
    return district_support
```

```python
def partisan_percentage(dataframe):
    ##This code shows you an example of how to calculate the population of a district, you can adapt it to count the partisan support.
    assignment = dataframe
    total_votes=0
    sum_votes=0
    support = 0
    sum_support = 0
    ovr_percentage = 0
    opposition = 0
    sum_opposition = 0
    district_support_per = []
    district_lean = 0
    for districts in (assignment.groupby(by="District")):
        #print(districts)
        for precinct in districts[1].iterrows():
            #print(precinct[1][0])
            if(pre_data.loc[pre_data.GEOID20==str(precinct[1][0]),'Total_2020_TotalAdj'].any()):        
                support += pre_data.loc[pre_data.GEOID20==str(precinct[1][0]),'Dem_2016-2021_Comp'].values[0]
                opposition += pre_data.loc[pre_data.GEOID20==str(precinct[1][0]),'Rep_2016-2021_Comp'].values[0]
                total_votes += pre_data.loc[pre_data.GEOID20==str(precinct[1][0]),'Total_2016-2021_Comp'].values[0]
                percentage  = support/total_votes
                percentage_opp = opposition/total_votes
        print("The percentage of District  "+str(districts[0])+ " that is Democratic is "+str(percentage * 100))
        print("The percentage of District  "+str(districts[0])+ " that is Republican is "+str(percentage_opp * 100))
        if percentage > percentage_opp:
            district_lean += 1
        district_support_per = district_support_per + [(str(districts[0]), str(percentage))]
        district_support_per.sort(key=sortSecond, reverse=True)
        sum_votes+=total_votes
        sum_support+=support
        ovr_percentage = sum_support/sum_votes
        total_votes=0
        support = 0
        opposition = 0
    print("Overall representation is "+str(ovr_percentage * 100))
    print(str(district_lean) + ' out of 40 districts are Democratic. Thats ' + str(district_lean/40 *100) + ' percent')
    print('We want ' + str(math.ceil(ovr_percentage *  40)) + ' districts')
    return(math.ceil(ovr_percentage *  40) - district_lean)
```
